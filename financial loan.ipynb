# Cross tab

- a cross is a statistical tool that displays the relationship between two or more categorical variables in a matrix formate.
- often used to understand the ditrubution of frequency of variables

```python
pd.crosstab(df["grade"],df["home_ownership"],values=df["loan_amount"],aggfunc="mean")
```
